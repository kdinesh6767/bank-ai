```python
from langchain.prompts import ChatPromptTemplate

# Create a ChatPromptTemplate with placeholders
chat_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a virtual assistant for a bank."),
        ("system", "How can I assist you today?"),
        ("user", "I would like to make a deposit."),
        ("system", "Great! How much would you like to deposit?"),
        ("user", "{deposit_amount}"),  # Placeholder for user input
        ("system", "Thank you. What is your account number?"),
        ("user", "{account_number}"),  # Placeholder for user input
        ("system", "Thank you for providing the information."),
    ]
)

# Create a dictionary to store user responses
user_responses = {}

# Iterate through messages
for role, content in chat_template.format_messages(deposit_amount="", account_number=""):
    # Print the message
    print(f"{role.capitalize()}: {content}")

    # If the role is "user," simulate user typing and store the response
    if role == "user":
        # Extract the key from {key} in the content
        key = content.strip("{}")

        # Prompt the user for a response using the extracted key
        user_responses[key] = input(f"Your response for {key}: ")

# Print the final formatted messages with user responses
for role, content in chat_template.format_messages(**user_responses):
    print(f"{role.capitalize()}: {content}")
```
